```python
var = 'age'
titanic.loc[titanic[var] >= titanic[var].mean(), :].head()   # 로 하면 변수 바꿔도 1줄만 수정하면 됨
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
var = 'total_bill'
print(tips[var].median())
tips.loc[tips[var] <= tips[var].median(),:]
```

```python
tips.loc[tips['day'] == 'Sun', :].head()
```

```python
tips.loc[(tips['tip'] > tips['tip'].mean()) & (tips['time'] == 'Dinner'), :].head()
```

```python
tips[(tips['tip'] > tips['tip'].mean()) & (tips['time'] == 'Dinner')][['tip','day']]    # loc과 그냥 추출 차이! 그냥 추출은 [행조건, 열조건] 형식 안돼서 밖으로 열 빼서 선택해야
```
